```python
import os
from langchain.llms import HuggingFacePipeline
from langchain.chains import LLMChain
from langchain.prompts import PromptTemplate
from transformers import AutoTokenizer, AutoModelForCausalLM, pipeline
```

```python
# Define a simple prompt template
prompt_template = PromptTemplate(
    input_variables=["message"],
    template="Say 'Hello, {message}!' in an enthusiastic way."
)
```

```python
# Load a small open-source model
# You can replace this with any model from Hugging Face Hub that fits your resources
model_name = "TinyLlama/TinyLlama-1.1B-Chat-v1.0"  # A small model that can run on CPU
```

```python
# Load the model and tokenizer
tokenizer = AutoTokenizer.from_pretrained(model_name)
model = AutoModelForCausalLM.from_pretrained(model_name)
```

```python
# Create a text generation pipeline
text_generation_pipeline = pipeline(
    "text-generation",
    model=model,
    tokenizer=tokenizer,
    max_new_tokens=100,
    temperature=0.7
)

# Create a LangChain wrapper around the Hugging Face pipeline
llm = HuggingFacePipeline(pipeline=text_generation_pipeline)

# Create a chain
chain = LLMChain(llm=llm, prompt=prompt_template)

# Run the chain
response = chain.run("Hello, How are you?")
print(response)
```
